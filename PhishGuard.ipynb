```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split

vocab_size = 3000
max_length = 30

tokenizer = tf.keras.preprocessing.text.Tokenizer(
    num_words=vocab_size,
    oov_token="<OOV>"
)
tokenizer.fit_on_texts(df['text'])

sequences = tokenizer.texts_to_sequences(df['text'])
padded = tf.keras.preprocessing.sequence.pad_sequences(
    sequences,
    maxlen=max_length,
    padding='post',
    truncating='post'
)

labels = df['label'].values

X_train, X_test, y_train, y_test = train_test_split(
    padded, labels, test_size=0.1, random_state=42
)

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 32, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

history = model.fit(
    X_train, y_train,
    epochs=20,
    validation_data=(X_test, y_test),
    verbose=1
)

print(f"\n최종 검증 정확도: {max(history.history['val_accuracy']):.4f}")
```

```python
import json

# TFLite 변환
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("phishing_model.tflite", "wb") as f:
    f.write(tflite_model)

print(f"모델 크기: {len(tflite_model) / 1024:.1f} KB")

# vocab 저장
vocab_data = {
    "word_index": tokenizer.word_index,
    "max_length": max_length,
    "vocab_size": vocab_size
}
with open("vocab.json", "w", encoding="utf-8") as f:
    json.dump(vocab_data, f, ensure_ascii=False)

print("저장 완료!")
```

```python
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()

def predict(text):
    seq = tokenizer.texts_to_sequences([text])
    padded_seq = tf.keras.preprocessing.sequence.pad_sequences(
        seq, maxlen=max_length, padding='post'
    )
    input_data = np.array(padded_seq, dtype=np.float32)

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()
    output = interpreter.get_tensor(output_details[0]['index'])
    return output[0][1]

test_cases = [
    # 명확한 피싱
    ("[Web발신] 계좌가 정지되었습니다 즉시 인증하세요", "피싱"),
    ("긴급 본인인증 필요 클릭하세요", "피싱"),
    ("국세청 세금 미납 즉시 납부하세요", "피싱"),
    ("택배 미수령 반송됩니다 주소 확인하세요", "피싱"),
    ("당첨되셨습니다 개인정보를 입력하세요", "피싱"),
    # 사회공학적 피싱 (친근하게 시작)
    ("안녕하세요 고객님 계좌가 정지되었습니다 즉시 인증하세요", "피싱"),
    ("고객님 중요한 안내사항입니다 계좌 확인 필요", "피싱"),
    ("안녕하세요 보안 강화 조치로 인증이 필요합니다", "피싱"),
    # 명확한 정상
    ("안녕하세요 내일 회의 있습니다", "정상"),
    ("신한카드 승인 스타벅스 5500원", "정상"),
    ("CJ대한통운 배송 완료 문 앞에 놓았습니다", "정상"),
    ("인증번호는 938271 입니다 타인에게 알리지 마세요", "정상"),
    ("오늘 저녁 같이 먹을까요", "정상"),
]

print("\n=== 모델 검증 ===")
correct = 0
for text, expected in test_cases:
    score = predict(text)
    result = "피싱" if score > 0.5 else "정상"
    is_correct = result == expected
    if is_correct:
        correct += 1
    emoji = "✓" if is_correct else "✗"
    print(f"{emoji} 예상: {expected} | 점수: {score:.3f} | 판정: {result} | {text[:35]}")

print(f"\n정확도: {correct}/{len(test_cases)} ({correct/len(test_cases)*100:.1f}%)")
```
